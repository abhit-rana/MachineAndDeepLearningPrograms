```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.metrics import accuracy_score
```

```python
# before that we i need to from the dataset of complete X and complete Y

X_train = np.concatenate((dC0, dC1, dC2, dC3, dC4, dC5, dC6, dC7, dC8, dC9))
X_test = np.concatenate((dCt0, dCt1, dCt2, dCt3, dCt4, dCt5, dCt6, dCt7, dCt8, dCt9))

Y_train = np.concatenate((np.array([0 for i in range(5000)]), np.array([1 for i in range(5000)]), np.array([2 for i in range(5000)]), np.array([3 for i in range(5000)]), np.array([4 for i in range(5000)]), np.array([5 for i in range(5000)]), np.array([6 for i in range(5000)]), np.array([7 for i in range(5000)]), np.array([8 for i in range(5000)]), np.array([9 for i in range(5000)])))
Y_test = np.concatenate((np.array([0 for i in range(1000)]), np.array([1 for i in range(1000)]), np.array([2 for i in range(1000)]), np.array([3 for i in range(1000)]), np.array([4 for i in range(1000)]), np.array([5 for i in range(1000)]), np.array([6 for i in range(1000)]), np.array([7 for i in range(1000)]), np.array([8 for i in range(1000)]), np.array([9 for i in range(1000)])))
```

```python
# without reduction of dimensions

clfwr = LDA()
clfwr.fit(X_train, Y_train)
```

```python
Y_pred = clfwr.predict(X_test)
print("accuracy:", accuracy_score(Y_test, Y_pred))
#############################
```
